```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x)

def f_prime(x):
    return -np.sin(x)

def f_prime_numerical(f,x,dx):
    return (f(x+dx)-f(x))/dx

def error(f,x,dx):
    return np.abs(f_prime_numerical(f,x,dx)-f_prime(x))

def log_error(f,x,dx):
    return np.log10(error(f,x,dx))

def log_dx(dx):
    return np.log10(dx)

def plot_error(f,x,dx):
    plt.plot(log_dx(dx),log_error(f,x,dx))
    plt.xlabel('log10(dx)')
    plt.ylabel('log10(Error)')
    plt.title('log10(Error) vs log10(dx)')
    plt.show()

plot_error(f,np.pi/4,np.logspace(-16,0,100))
```

```python
def central_difference(f,x,dx):
    return (f(x+dx)-f(x-dx))/(2*dx)

def error(f,x,dx):
    return np.abs(f_prime(x)-central_difference(f,x,dx))

def log_error(f,x,dx):
    return np.log10(error(f,x,dx))

def log_dx(dx):
    return np.log10(dx)

def plot_error(f,x,dx):
    plt.plot(log_dx(dx),log_error(f,x,dx))
    plt.xlabel('log10(dx)')
    plt.ylabel('log10(error)')
    plt.title('log10(Error) vs log10(dx)')
    plt.show()

plot_error(f,np.pi/4,np.logspace(-16,0,100))
```
