```python
# Homework

# Undergraduate student:
# A person throws a ball upwards at a speed of 10 m/s, please calculate the trajectory using Runge-Kutta method
# and plot height (z) vs time (t).

import numpy as np
import matplotlib.pyplot as plt

tend=3
h=.001
y=[0.,10.]
t=0

def f(t, y):
    f=np.zeros(2)
    f[0]=y[1]
    f[1]=-9.81
    return f
    
def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1=h*f(t,y)
    k2=h*f(t+0.5*h,y+0.5*k1)
    k3=h*f(t+0.5*h,y+0.5*k2)
    k4=h*f(t+h,y+k3)
    y=y+(k1+2*k2+2*k3+k4)/6
    return y

time=[]
yp=[]
while(t<tend):
    if((t+h)>tend):
        h=tend-t
    y = rk4(t,h,y,2)
    if(y[0]<0):
        y[0]=0
    yp=np.append(yp,y[0])
    t = t+h
    time=np.append(time,t)
plt.title('Ball Trajectory')
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
plt.plot(time,yp)
plt.show()
```
