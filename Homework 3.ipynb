```python
#Recursion method
import time
start = time.time()
x=[]
ti=[]
n1=30+1
nn=np.arange(n1)
for i in nn:
    def fib2(i):      
        if i == 0: return 0
        elif i == 1: return 1
        else: return fib2(i-1)+fib2(i-2)
    ti.append(time.time() - start)
    x.append(fib2(i))
    
print(x)
print(ti)
print("--- %s seconds ---" % (time.time() - start))
```

```python
#Recursion continued
x_axis=np.log10(x[1:])
y_axis=np.log10(ti[1:])
print(y_axis)
print(x_axis)
plt.plot(x_axis,y_axis)
plt.xlabel("log10(n)")
plt.ylabel("log10(time)")
```

```python
#Matrix method
start_time = time.time()
def arrpow(arr, n):
    yarr=arr
    if n<1:
        print('n needs to be larger than 1')
    if n==1:
        return arr
    yarr = arrpow(arr, n//2)
    yarr = [[yarr[0][0]*yarr[0][0]+yarr[0][1]*yarr[1][0],yarr[0][0]*yarr[0][1]+yarr[0][1]*yarr[1][1]],
            [yarr[1][0]*yarr[0][0]+yarr[1][1]*yarr[1][0],yarr[1][0]*yarr[0][1]+yarr[1][1]*yarr[1][1]]]
    if n%2:
        yarr=[[yarr[0][0]*arr[0][0]+yarr[0][1]*arr[1][0],yarr[0][0]*arr[0][1]+yarr[0][1]*arr[1][1]],
            [yarr[1][0]*arr[0][0]+yarr[1][1]*arr[1][0],yarr[1][0]*arr[0][1]+yarr[1][1]*arr[1][1]]]
    return yarr

def fib5(n):
    arr= [[1,1],[1,0]]
    f=arrpow(arr,n-1)[0][0]
    return f


print(fib5(40))
print("--- %s seconds ---" % (time.time() - start_time))
```
