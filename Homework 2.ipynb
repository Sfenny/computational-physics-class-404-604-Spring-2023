```python
# Homework 2: Write a program to plot
# 𝑓(𝑥)=𝜋/2+2∗(𝑠𝑖𝑛(𝑥)+𝑠𝑖𝑛(3𝑥)/3+𝑠𝑖𝑛(5𝑥)/5+....𝑠𝑖𝑛(𝑛𝑥)/𝑛)
# from - 𝜋 to  𝜋 Choose n = 11, 21 and 31 and plot all these three curves in one figure. Please use numpy library and matplotlib.

import numpy as np
import matplotlib.pyplot as plt

a = np.arange(1., 12., 2.)
b = np.arange(1., 22., 2.)
c = np.arange(1., 32., 2.)
g = np.arange(-np.pi,np.pi,0.0001)
x = np.outer(g,a)
y = np.outer(g,b)
z = np.outer(g,c)
x_ticks = [-np.pi,-np.pi/2,0,np.pi/2,np.pi]
xx = np.sin(x)/a
yy = np.sin(y)/b
zz = np.sin(z)/c

xxx = np.sum(xx, axis=1)
yyy = np.sum(yy, axis=1)
zzz = np.sum(zz, axis=1)

n_1= np.pi/2+(2*xxx)
n_2= np.pi/2+(2*yyy)
n_3= np.pi/2+(2*zzz)

plt.plot (g,n_1)
plt.plot (g,n_2)
plt.plot (g,n_3)
plt.show ()
```
