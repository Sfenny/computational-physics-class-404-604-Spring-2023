## Homework 8:

Solve the diffusion equation
\begin{equation}
U_t = D*U_{xx} 
\end{equation}
Where,
\begin{equation}
D=1,\qquad 0<x<1 ,\qquad  0<t<0.1 
\end{equation}

The initial condition is
\begin{equation}
U(x,0)=\frac{cos(13x)+1}{2}
\end{equation}
and the boundary conditions are
\begin{equation}
U(0,t)=1, \qquad U(1,t)=0
\end{equation}



```python
from numpy import *
import numpy as np
import matplotlib.pylab as p
from mpl_toolkits.mplot3d import Axes3D 

Nx = 101;       tend=.1;     Dt=0.0001;      Dtout=0.001;     Dx = .02;                                                                                                                 
T = zeros((Nx,2),float);  Tp = zeros((Nx,int(tend/Dtout)+1),float)  
# T[i,0] is the temperature at position i and old time, T[i, 1] is at the new time                                    

for ix in range (1, Nx - 1):  T[ix, 0] = 0.5*(np.cos(13*ix)+1);               # Initial T
T[0,0] = 1.0 ;   T[0,1] = 0.0                                                 # 1st & last T = 0
T[Nx-1,0] = 0.0; T[Nx-1,1] = 0.0
cons = Dt/(Dx*Dx);                                                            # constant=D*dt/Dx^2
m = 1                                                                         # counter

toutn=0
t=0.
while t < tend:
    for ix in range (1, Nx - 1):
        T[ix, 1] =T[ix, 0]+cons*(T[ix+1, 0]-2*T[ix, 0]+T[ix-1, 0])
    t+=Dt
    for ix in range (1, Nx - 1):  T[ix, 0] = T[ix, 1]

    if t >= toutn*Dtout:
        p.plot(T)
        Tp[:,toutn] = T[:,1]
        toutn += 1
    

        
x = list(range(0, Nx))
y = list(range(0, int(tend/Dtout)+1))
X, Y = p.meshgrid(x, y)

def functz(Tpl):
    z = Tpl[X, Y]
    return z

Z = functz(Tp)  
fig=p.figure()
ax = fig.add_subplot(projection='3d')                                            
ax.plot_wireframe(X, Y, Z, color = 'r')                    
ax.set_xlabel('Position')                                     
ax.set_ylabel('time')
ax.set_zlabel('Temp')
p.show()                               
print("finished")       
```
